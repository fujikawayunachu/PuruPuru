```python
import numpy as np
import matplotlib.pyplot as plt

# xの範囲を指定（例えば -10 から 50）
x = (np.linspace(0, 100, 100))
y = (0.9 ** x) * np.sin(x)*0.5 +1

# グラフを描画
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='y = 0.9833^x * sin(x)', color='teal')
plt.title('y = 0.9833^x * sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='gray', linewidth=0.8)
plt.axvline(0, color='gray', linewidth=0.8)
plt.legend()
plt.show()
```
